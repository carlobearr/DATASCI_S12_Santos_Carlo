```python
import requests
import json
import pandas
from newsapi import NewsApiClient
# import os
# from dotenv import load_dotenv

# load_dotenv()

# NEWSAPI_KEY = os.getenv('NEWSAPI_KEY')

# Init
newsapi = NewsApiClient(api_key=NEWSAPI_KEY)
```

```python
all_articles = newsapi.get_everything(
    sources='cnn',
    from_param='2021-03-11',
    to='2021-03-12',
    language='en')


all_articles
```

```python
results= all_articles['articles']
results
```

```python
results[0]
```

```python
import datetime
news_json = []


for i in range(0, len(results)):
    date_time_str = results[i]['publishedAt']
    date_time_obj = datetime.datetime.strptime(date_time_str, '%Y-%m-%dT%H:%M:%SZ')
    date= str(date_time_obj.date())
    title=results[i]['title']
    fullarticle=results[i]['content']
    author=results[i]['author']
    news_json.append({
        "date": date,
        "title": title,
        "full article": fullarticle,
        "author": author
    })
    
news_json
```

```python
with open('march11_12news.json', 'w') as json_file:
  json.dump(news_json, json_file)

print(json.dumps(news_json, indent = 4))
```
